# Compute by hand the complex multiplication (5 – 3j) ∗ (4 + 2j). Reform the same operation using python

# Computation by hand


(5 - 3j) * (4 + 2j)

= 5 * 4 + 5 * 2j - 3j * 4 - 3j * 2j

= 20 + 10j - 12j - 6j^2

= 20 + (10j - 12j) - 6j^2

= 20 - 2j - 6j^2

= 20 - 2j + 6 (Since j^2 = -1)

= 26 - 2j
