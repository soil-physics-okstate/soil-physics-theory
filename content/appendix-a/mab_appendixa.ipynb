```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4*np.pi, 4*np.pi, 500)
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(-4*np.pi, 4*np.pi, 500)
y = np.cos(x)
plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(-4*np.pi, 4*np.pi, 500)
y = np.tan(x)
plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(-4*np.pi, 4*np.pi, 1000)
y = np.exp(-np.abs(x))
plt.plot(x, y, color = 'green')
plt.show()
```
