**5.3. Compute the matric water potential for water in a saturated pore 10 μm in diam- ◦ eter, at 20 C. The surface tension and liquid density at this temperature are 72.75×10–3 Nm–1 and1027.8kgm–3,respectively.**

```python
import math
import numpy as np
r=5/1000000 #μm
temp=20 #c
beta=0
ST=72.75*10**-3. #Nm–1
LD= 1027.8 #kg/m3
water_potential= (2*ST*np.cos(beta))/ (LD*r)
print("matric_water_potential:","%.2f" %water_potential, "kpa")
```
